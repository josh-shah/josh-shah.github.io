```python
import pandas as pd

#Import TrendReq to connect to Google
from pytrends.request import TrendReq

#Use exact keyword codes (found in https://colab.research.google.com/drive/1lTSIMgRJMBXQMXSvKt-CDEE_zunOsZEF#scrollTo=VC_k3IiFTwXx) to avoid ambiguity, trying Team names led to inaccurate results. 
list1=['/m/050fh','/m/0xbm','/m/04ltf	','/m/023fb	','/m/01634x'] #Man U, Arsenal, Liverpool, Chelsea, Man City

pytrends1=TrendReq()  

#build payload for both the lists and pass parameters to filter the data for trend comparison based on our requirements
pytrends1.build_payload(list1,timeframe='today 12-m')

df1=pytrends1.interest_by_region(resolution='COUNTRY', inc_low_vol=True, inc_geo_code=True)
```
