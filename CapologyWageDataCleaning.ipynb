```python
import pandas as pd
```

```python
df = pd.read_csv('all_premier-league_2022-2023_last_updated_15112022.csv')
```

```python
#The Capology Scrape left us with totals of each Club in rows, so we need to drop these'
df = df.drop([23,24,52,53,81,82,107,108,132,133,159,160,185,186,214,215,240,241,269,270,297,298,326,327,351,352,384,385,414,415,446,447,472,473,499,500,525,526,552])
```

```python
#Drop columns that we don't need
df = df.drop(['Gross P/Y(IN GBP)','Gross(IN GBP)','Expiration Jun 30, 2023Jun 30, 2024Jun 30, 2025Jun 30, 2026Jun 30, 2027','Length 1-yr2-yrs3-yrs4-yrs5-yrs','GrossRemaining(IN GBP)','ReleaseClause(IN GBP)','Status ReserveStarter','Active'],axis=1)
```

```python
df
```

```python
#Rename column to more readable name
df.rename(columns={'Gross P/W(IN GBP)': 'weekly_wage'} , inplace=True)
```

```python
#Drop anyone with unknown wages
df = df[df.weekly_wage != '£0.00']
```
